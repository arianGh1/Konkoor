```python
import pandas as pd
import pickle
import numpy as np
```

```python
with open("dic_gc1","rb") as f:
    dic_riazi = pickle.load(f)
with open("dic_gc3","rb") as f:
    dic_tajrobi = pickle.load(f)
with open("dic_gc5","rb") as f:
    dic_ensani = pickle.load(f)
with open("dic_gc7","rb") as f:
    dic_honar = pickle.load(f)
with open("dic_gc9","rb") as f:
    dic_zaban = pickle.load(f)
```

```python
df_riazi = pd.DataFrame.from_dict(dic_riazi.values())
df_tajrobi = pd.DataFrame.from_dict(dic_tajrobi.values())
df_ensani = pd.DataFrame.from_dict(dic_ensani.values())
df_honar = pd.DataFrame.from_dict(dic_honar.values())
df_zaban = pd.DataFrame.from_dict(dic_zaban.values())
```

```python
frames = [df_riazi,df_tajrobi,df_ensani,df_honar,df_zaban]
df = pd.concat(frames,axis=0)
```

```python
df = df.drop_duplicates()
df = df.reset_index(drop=True)
```

```python
df = df[~(pd.isnull(df["نام و نام خانوادگی"]))]
```

```python
df
```

```python
df["رشته مدرسه"] = df["gc"].apply(lambda x: "ریاضی" if x==1 else x)
df["رشته مدرسه"] = df["رشته مدرسه"].apply(lambda x: "انسانی" if x==5 else x)
df["رشته مدرسه"] = df["رشته مدرسه"].apply(lambda x: "زبان" if x==9 else x)
del df["gc"]
```

```python
df
```

```python
df["سال"] = df["year"]
del df["year"]
```

```python
df["سال"] = df["سال"]+1300
```

```python
df.to_csv("result2.csv",index=False)
```
